```python
import math
def euclidean_distance(point1, point2):
    return math.sqrt(sum((p2 - p1) ** 2 for p1, p2 in zip(point1, point2)))
def get_coordinates():
    x = float(input("Enter x coordinate: "))
    y = float(input("Enter y coordinate: "))
    z = float(input("Enter z coordinate: "))
    return x, y, z

def main():
    print("Enter coordinates for point 1:")
    point1 = get_coordinates()
    print("\nEnter coordinates for point 2:")
    point2 = get_coordinates()

    distance = euclidean_distance(point1, point2)

    print("\nEuclidean Distance between the two points:", distance)

if __name__ == "__main__":
    main()
```

```python
from sklearn.datasets import load_iris
from scipy.spatial.distance import cdist

iris = load_iris()
data = iris.data

distance_matrix = cdist(data, data, metric='euclidean')
print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
X = iris.data
y = iris.target
class_0_indices = np.where(y == 0)[0]
class_1_indices = np.where(y == 1)[0]
indices = np.concatenate([class_0_indices, class_1_indices])
X_filtered = X[indices]
y_filtered = y[indices]


feature1 = X_filtered[:, 0]
feature2 = X_filtered[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(feature1[class_0_indices], feature2[class_0_indices], c='r', label='Class 0')
plt.scatter(feature1[class_1_indices], feature2[class_1_indices], c='b', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Two Classes')
plt.legend()
plt.grid(True)
plt.axvline(x=(np.max(feature1[class_0_indices]) + np.min(feature1[class_1_indices])) / 2, color='k', linestyle='--')

plt.show()
```
